### Q8: What is the average number of properties in council tax band D? [6]

```python
#Use mean function to calc average
mcnltx_df['Count - Council Tax Band D'].mean()
```

### Q9: What is the total number of properties in Manchester? [7]

```python
#Add all bands together to get total for each area
TotalPropsMan = mcnltx_df['Count - Council Tax Band A'] + mcnltx_df['Count - Council Tax Band B'] + mcnltx_df['Count - Council Tax Band C'] + mcnltx_df['Count - Council Tax Band D'] + mcnltx_df['Count - Council Tax Band E'] + mcnltx_df['Count - Council Tax Band F'] + mcnltx_df['Count - Council Tax Band G']+ mcnltx_df['Count - Council Tax Band H']
#store in variable and print
TotalPropsMan
```

```python
#Sum the values calculated and print total
TotalProps = TotalPropsMan.sum()
TotalProps
```

### Q10: Compute total number of properties in each ward and assign it to a new column called `total` [8]

```python
#Add this new computed data into total column in dataset
mcnltx_df['Total'] =  mcnltx_df['Count - Council Tax Band A'] + mcnltx_df['Count - Council Tax Band B'] + mcnltx_df['Count - Council Tax Band C'] + mcnltx_df['Count - Council Tax Band D'] + mcnltx_df['Count - Council Tax Band E'] + mcnltx_df['Count - Council Tax Band F'] + mcnltx_df['Count - Council Tax Band G']+ mcnltx_df['Count - Council Tax Band H']
mcnltx_df.head()
```

### Q11: Show all the wards containing more than 3.5% of all properties in Manchester [9]

```python
#calculate using totals to find percentage and see if greater than 3.5%
mcnltx_df.loc[ ((mcnltx_df['Total'] / TotalProps)  > 0.035)]
```

### Q12: Create a new column called `above_average` which should contain a `'Y'` if the number of properties is above the average for Manchester or an `'N'` otherwise [9]

```python
#Find the average props in each area
AverProps = mcnltx_df['Total'].mean()
#get total - both variables make where statement more readable
tot = mcnltx_df['Total']
#use np.where to input Y or N for above average criteria
mcnltx_df['Above Average'] = np.where (tot > AverProps, 'Y', 'N')
mcnltx_df.head()
```

### Q13: Make the ward name the new row index of the dataframe [5]

```python
#Set index to name - display changes
mcnltx_df.set_index('Name', inplace=True)
mcnltx_df.head()
```

### Q14: Create a new row whose name is "Manchester" and include in it the totals for every band [8]

```python
#Create new row called Manchester and produce totals for each column
mcnltx_df.loc['Manchester']= mcnltx_df.sum()
mcnltx_df.tail()
```

### Q16: Show the name of the wards that have more than 5,000 properties in Band A [5]

```python
#Store all band A's in Variable
BdA = mcnltx_df['Count - Council Tax Band A']
#Use where to find all above 5000 and drop any that are not applicable 
BdA.where(BdA > 5000).dropna()
```
